# EDA for the world population
## description
This Jupyter notebook contains an exploratory data analysis (EDA) of a world population dataset. The dataset includes information on population counts and percentages across various countries and continents over several decades.

### Importing Necessary Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### Reading the CSV File into a DataFrame

```python
path = "./world_population.csv"
df = pd.read_csv(path)
df
```

### Getting an Overview of the DataFrame Structure

```python
df.info()
```

### Generating Descriptive Statistics

```python
df.describe()
```

### Checking for Missing Values

```python
df.isnull().sum()
```

### Counting Unique Values in Each Column

```python
df.nunique()
```

### Sorting Data by a Specific Column

```python
df.sort_values(by='World Population Percentage' , ascending=False).head(10)
```

### Dropping Non-Numerical Columns and Calculating Correlation

```python
num_df = df.drop(columns=['CCA3', 'Country', 'Capital', 'Continent'])
num_df.corr()
```

### Visualizing the Correlation Matrix

```python
plt.rcParams['figure.figsize'] = (20, 7)
sns.heatmap(num_df.corr(), annot=True , fmt='.2f')
plt.show()
```

```python
df.columns
```

### Grouping Data by Continent and Calculating Mean Population

```python
num_continent = df.drop(['CCA3', 'Country', 'Capital'], axis=1)
display(num_continent.groupby('Continent').mean().sort_values(by='2022 Population', ascending=False))
```

### Averaging Population by Decade and Sorting by 2022 Population

```python
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by="2022 Population",ascending=False)
df2
```

### Transposing the Grouped DataFrame

```python
df3 = df2.transpose()
df3
```

### Plotting Population Trends

```python
df3.plot()
```

### Creating a Boxplot of the DataFrame

```python
df.boxplot(figsize=(20,10))
```
